```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("F:\ML\Machine Learning\CSV file\sonar_dataset.csv",header=None)
df.sample(10)
```

```python
X = df.drop(60,axis='columns')
y = df[60]
y.head()
```

```python
y = pd.get_dummies(y,drop_first=True)
y
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(64,input_dim=60,activation='relu'), 
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(16,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
model.evaluate(X_train,y_train)
```

```python
y_pred = np.round(model.predict(X_test).reshape(-1))
y_pred[:10]
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Pred 0','Pred 1'], yticklabels=['True 0','True 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(f'Confusion Matrix (Accuracy = {accuracy:.2f})')
plt.show()
```

```python
y_prob = model.predict(X_test).reshape(-1)

sns.histplot(y_prob, kde=True)
plt.title("Distribution of Predicted Probabilities")
plt.xlabel("Predicted Probability")
plt.ylabel("Count")
plt.show()
```

```python
modeld.evaluate(X_test, y_test)
```
