```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('placement.csv')
```

```python

print(df.shape)
df.head()
```

```python
X = df[['cgpa', 'resume_score']].values   
y = df['placed'].astype(int).values 
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

```python
scaler = StandardScaler()
X_sca = scaler.fit_transform(X)
```

```python
model = Perceptron()
model.fit(X_sca, y)
```

```python
plot_decision_regions(X=X_sca, y=y, clf=model, legend=2)

plt.xlabel('CGPA (standardized)')
plt.ylabel('Resume Score (standardized)')
plt.title('Perceptron Decision Boundary (Standardized Data)')
plt.show()
```
