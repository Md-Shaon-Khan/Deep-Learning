```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("F:\Deep Learning\Perceptron\placement.csv")
df.shape
```

```python
df.head()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
X = df[['cgpa', 'resume_score']].values  
y = df['placed'].astype(int).values 
```

```python
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
stan = StandardScaler()
model = Perceptron()
```

```python
X_sca = stan.fit_transform(X)
```

```python
model.fit(X_sca,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X=X_sca,y=y,clf=model,legend=2)
```
